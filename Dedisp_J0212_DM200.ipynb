```python
#A script that plots the de-dispersion plan for a suspected binary pulsar system.
#Dispersion is an index of refraction that is frequency dependent resulting from the propagation
#of light through the interstellar medium, this needs to be corrected for before data can be analyzed
#this script plots the plan for this correction. Dispersion measure (DM) has units of parsec
#per square cm, and smearing has units of milliseconds.
#Smearing is the amount that the pulse signal is smeared or broadened in each channel.

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter

# Data for plotting
t = [0.0, 12.5, 25, 37.5, 50, 75, 100, 125, 150, 175, 200]
s = [0.004, 0.01, 0.004, 0.01, 0.004, 0.0364, 0.004, 0.0364, 0.004, 0.0364, 0.004]

def extendlist(l):
    master = []
    for i in range(len(l)-1):
        x = np.linspace(l[i], l[i+1], 50)
        master.extend(x)
    return master
#defining the function extendlist, for looping for i in the range of the length of l-1
#defining the variable x as evenly spaced numbers from l[i], l[i+1] and 50
#extending x by appending all the items from the iterable

t = extendlist(t)
s = extendlist(s)

#redefining the variables't' and 's' to extended lists of their values

fig, ax = plt.subplots()
ax.semilogy(t, s)
# create a plot and subplots defined as 'fig, ax'
#create a log plot for each axis

ax.set(xlabel='DM / pc cm-3', ylabel='Smearing / ms', title='De-dispersion Plan')
plt.xlim((0,200))
plt.ylim((0.001, 1))
#setting the axis labels, and the upper and lower limits of the axes

plt.savefig('Coherent de-disp plot.png', dpi=125)
plt.show()
#saving and showing the image
```
