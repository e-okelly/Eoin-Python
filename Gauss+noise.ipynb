```python
#A script that generates and plots a gaussian distribution

import numpy as np
import matplotlib.pyplot as plt
a = np.loadtxt('file.txt')
#loading text from the file 'file.txt'

print(a)

x = a[0:,0]
y = a[0:,1]

print(x);print(y)
#defining x and y as lists of the data from a

x = y = np.arange (-700,700,1)
#returning evenly spaced values for x and y between -700 and 700 in steps of 1

xx,yy = np.meshgrid (x,y)
#returning coordinate matrices from coordinate vectors for x and y 

z = np.hypot (xx,yy)
#assigning the value of the hypotenuse of a right angle triangle of xx,yy to the variable z

gauss_image = np.exp(-z*z/(200*200))
#calculating the exponential of all elements in -z*z/(200*200) and assigning them to the variable gauss_image

plt.plot(gauss_image)
plt.show()
#plotting and showing gauss_image
```

```python
#a script that generates and plots noise

import numpy as np
pure = np.linspace(-1, 1, 200)
#returns evenly spaced numbers over the specified interval of -1 to 1 up to 200 and assigns it
# to the variable 'pure'

noise = np.random.normal(0, 1, pure.shape)
#draws random samples from a normal gaussian distribution and assigns them to the variable 'noise'

signal = pure + noise
#assigns the two variables 'pure' + 'noise' to the variable 'signal'

plt.plot(pure + noise + signal)
plt.show
#plotting and showing 'pure' + 'noise' + 'signal'
```
