```python
#This is a script to plot light curves (a plot of changing brightness - orbital phase vs differential magnitude)
#It takes the data file, reads through it line by line using a for loop and then plots it
#The commented out lines below are if you want to plot uncertainty lines rather than the data points

import astropy 
import matplotlib.pyplot as plt
import matplotlib.pylab as plb

filename = "J1147_LC_data-2.prn"

f = open("J1147_LC_data-2.prn", "r")

for A in f:
    x = float(A[30:40])
    y = float(A[24:30])
    
    #y_error = float(A[74:105])
    #y_uncert = y_error, y_error
    
    x2 = x
    y2 = y
    
    plt.scatter(x,y, color='black', s=5, marker=".")
    #plt.scatter(x2,y2, color='black', s=5, marker=".")
    
    plt.ylim(0.3,-0.3)
    #plt.errorbar(x2, y2, color='blue', marker=",", elinewidth = 1, fmt = 'o')
    
    plt.xlabel('Orbital Phase')
    plt.ylabel('Differential Magnitude')
    plt.savefig('J1147_light_curve.png',dpi=500)
    
    plt.show
    
    #print(x,y)
```
